# Predicting Price Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
df['Price']=''
```

```python
x=df[[ 'Rating', 'Specs', 'MODEL','RAM',
       'ROM']]
```

```python
y=df['Mrp']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
lr=lr.fit(x_train,y_train)
```

```python
prediction=lr.predict(x)
```

```python
prediction
```

```python
lr.score(x_test,y_test)
```

# Predicting Price Using Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf=RandomForestRegressor()
ref=reg_rf.fit(x_train,y_train)
```

```python
prediction1=ref.predict(x)
```

```python
prediction1
```

```python
ref.score(x_test,y_test)
```

```python
df['Price']=prediction1
```
